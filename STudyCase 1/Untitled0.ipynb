```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, numpy as np
!mkdir data
!wget -P data/ https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/jakarta.csv
!wget -P data/ https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/depok.csv
!wget -P data/ https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/bogor.csv
!wget -P data/ https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/tangerang.csv
!wget -P data/ https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/bekasi.csv
dfJakarta = pd.read_csv("data/jakarta.csv")
dfDepok = pd.read_csv("data/depok.csv")
dfBogor = pd.read_csv("data/bogor.csv")
dfTangerang = pd.read_csv("data/tangerang.csv")
dfbekasi = pd.read_csv("data/bekasi.csv")
```

# Bekasi

```python
# Melihat data terbalik dimana kolom vertikal baris horizontal
dfbekasi.transpose()
```

```python
# Karena dari data, terdapat beberapa kolom yang tidak bernilai "unamed:14-25" termasuk variabel "created_at" dan "URL" maka
# Drop 14 kolom tersebut
columnsdrop = ["Unnamed: 14",
           "Unnamed: 15",
           "Unnamed: 16",
           "Unnamed: 17",
           "Unnamed: 18",
           "Unnamed: 19",
           "Unnamed: 20",
           "Unnamed: 21",
           "Unnamed: 22",
           "Unnamed: 23",
           "Unnamed: 24",
           "Unnamed: 25",
           "created_at"]
dfbekasi = dfbekasi.drop(columns=columnsdrop)
dfbekasi
```

```python
# cek info data
dfbekasi.info()
```

```python
# Menghapus semua baris yang memiliki missing value di semua kolom, karena tidak bernilai
dfbekasi_2 = dfbekasi.dropna(how = 'all')

# Menampilkan DataFrame hasil setelah penghapusan
print(dfbekasi_2)
```

```python
# Mengecek tipe variabel
dfbekasi_2.info()
```

```python
# Cek nilai dari 3 kolom object
print(dfbekasi_2['sertifikat'].unique())
print(dfbekasi_2['hadap'].unique())
print(dfbekasi_2['lokasi'].unique())

```

```python
# Dari hasil diatas, ubah tipe data 'sertifikat', 'hadap', 'lokasi' ke tipedata 'category'
dfbekasi_2['sertifikat'] = dfbekasi_2['sertifikat'].astype('category')
dfbekasi_2['hadap'] = dfbekasi_2['hadap'].astype('category')
dfbekasi_2['lokasi'] = dfbekasi_2['lokasi'].astype('category')

# Mengecek tipe variabel
dfbekasi_2.info()
```

```python
# Data update
dfbekasi_2
```

### Missing Value

```python
# Cek missing value dari setiap variabel
dfbekasi_2.isnull().sum()
```

```python
# Cek missing value dengan persentase
(dfbekasi_2.isnull().sum()/len(dfbekasi_2)).to_frame('persentase missing')
```

```python
# cek missing value dengan gambar
sns.heatmap(dfbekasi_2.isnull(), cbar=False)
plt.title('Heatmap Missing Value')
plt.show()
```

```python
# isi mv dengan mean atau modus

dfbekasi_2["listrik"].fillna(dfbekasi_2["listrik"].quantile(0.5), inplace=True)
dfbekasi_2["KM"].fillna(dfbekasi_2["KM"].mode()[0], inplace=True)
dfbekasi_2["garasi"].fillna(dfbekasi_2["garasi"].mode()[0], inplace=True)
dfbekasi_2["carport"].fillna(dfbekasi_2["carport"].mode()[0], inplace=True)
dfbekasi_2["sertifikat"].fillna(dfbekasi_2["sertifikat"].mode()[0], inplace=True)
dfbekasi_2["hadap"].fillna(dfbekasi_2["hadap"].mode()[0], inplace=True)

# cek mv setelah diimputasi
dfbekasi_2.isnull().sum()
```

```python
# Mengecek duplikat data
print(dfbekasi_2.shape)
print("Total duplikat data: ", dfbekasi_2.duplicated().sum())
```
